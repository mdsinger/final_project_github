```python
# Initial imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data.csv dataset.
file_path = "Population-by-County-1850-2010.csv"
df = pd.read_csv(file_path)#, index_col=0)
df
```

```python
df.columns
```

```python
#Remove columns
df=df.drop(["1850", "1860", "1870","1880","1890", "1900", "1910", "1920","1930","1940", "1950", "1960","1970"], axis=1)
df
```

```python
df.dtypes
```

```python
#Rename columns
df.rename(columns={"1980":"1980 Census","1990":"1990 Census", "2000":"2000 Census", "2010":"2010 Census"}, inplace=True)
df.head()
```

```python
len(df)
```

```python
#Save new df
df.to_csv("data1.csv")
```

```python
#Merge the dataframes on the 'key' column
population_df = pd.merge(df, latest_df, on='County', how='outer')
population_df = pd.merge(population_df, predictions_df, on='County', how='outer')

# print the merged dataframe
population_df
```

```python
#Plot Bar Chart
ax = fp_df.plot(x='County', y=['1980 Census', '1990 Census', '2000 Census', '2010 Census', '2020 Census', '2022 Predictions'],
             kind='bar', figsize=(10, 6))
ax.set_xlabel('County')
ax.set_ylabel('Population')
ax.set_title('Population Census Data')

# Add labels for each individual bar
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
import seaborn as sns

# Create a dataframe 
data = {'County': ['Davis County', 'Salt Lake County', 'Utah County', 'Weber County'],
        '1980 Census': [146540, 619066, 218106, 144616],
        '1990 Census': [187941, 725956, 263590, 158330],
        '2000 Census': [238994, 898387, 368536, 196533],
        '2010 Census': [306479, 1029655, 516564, 231236],
        '2020 Census': [362679, 1185238, 659399, 262223],
        '2022 Predictions': [369948, 1186257, 702434, 269561]}
df = pd.DataFrame(data)

# Set 'County' as the index
df.set_index('County', inplace=True)


plt.figure(figsize=(10, 6))  # Adjust the width and height as needed

# Plot the heatmap
sns.heatmap(df, cmap='YlGnBu', annot=True, fmt=",d", cbar=True)
plt.title('Population by County and Year Heatmap')
plt.xlabel('Year')
plt.ylabel('County')
plt.show()
```
