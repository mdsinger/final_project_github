```python
# Initial imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data.csv dataset.
file_path = "Population-by-County-1850-2010.csv"
df = pd.read_csv(file_path)#, index_col=0)
df.head()
```

```python
df.columns
```

```python
#Remove columns
df=df.drop(["1850", "1860", "1870","1880","1890", "1900", "1910", "1920","1930","1940", "1950", "1960","1970"], axis=1)
df
```

```python
df.dtypes
```

```python
#Rename columns
df.rename(columns={"1980":"1980 Census","1990":"1990 Census", "2000":"2000 Census", "2010":"2010 Census"}, inplace=True)
df.head()
```

```python
len(df)
```

```python
#Save new df
df.to_csv("data1.csv")
```

```python
# Load the data.csv dataset.
file_path = "2010 2020 Census.csv"
latest_df = pd.read_csv(file_path)#, index_col=0)
latest_df.head()
```

```python
latest_df.columns
```

```python
#Create new Df
latest_df=latest_df[["County","2020 Census"]]
latest_df.head()
```

```python
len(latest_df)
```

```python
#Save new df
latest_df.to_csv("data2.csv")
```

```python
# Load the data.csv dataset.
file_path = "Predictions 2022.csv"
predictions_df = pd.read_csv(file_path)#, index_col=0)
predictions_df.head()
```

```python
#Create new Df
predictions_df=predictions_df[["County","2022"]]
predictions_df
```

```python
len(predictions_df)
```

```python
#Rename columns
predictions_df.rename(columns={"2022":"2022 Predictions"}, inplace=True)
predictions_df.head()
```

```python
#save file
predictions_df.to_csv("data3.csv")
```

```python
#Merge the dataframes on the 'key' column
population_df = pd.merge(df, latest_df, on='County', how='outer')
population_df = pd.merge(population_df, predictions_df, on='County', how='outer')

# print the merged dataframe
population_df
```

```python
population_df.to_csv("population.csv")
```

```python
population_df.dtypes
```

```python
population_df=population_df.replace(',',"",regex=True)
population_df=population_df.astype({"1980 Census":float,"1990 Census":float,"2020 Census":float, "2000 Census":float, "2010 Census":float, "2020 Census":float, "2022 Predictions":float })
population_df.dtypes
```
